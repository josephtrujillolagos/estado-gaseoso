medidas de tendencia central

```python
import statistics

# Datos de ejemplo
data = [10, 20, 30, 40, 50, 60, 70, 80, 90]

# Calcular la media aritmética
media = statistics.mean(data)
print(f"Media aritmética: {media}")

# Calcular la mediana
mediana = statistics.median(data)
print(f"Mediana: {mediana}")

# Calcular la moda
moda = statistics.mode(data)
print(f"Moda: {moda}")
```

medidas de dispersion

```python
import statistics

# Datos de ejemplo
data = [10, 20, 30, 40, 50, 60, 70, 80, 90]

# Calcular el rango
rango = max(data) - min(data)
print(f"Rango: {rango}")

# Calcular la varianza
varianza = statistics.variance(data)
print(f"Varianza: {varianza}")

# Calcular la desviación estándar
desviacion_estandar = statistics.stdev(data)
print(f"Desviación estándar: {desviacion_estandar}")
```

Actividad: Elaborar dataframes y comparar sus medidas de tendencia central para explicar las gráficas (Ref. Teoría cinética molecular de los gases)

```python
import pandas as pd

# Crear dos DataFrames de ejemplo
data1 = {'A': [100, 300, 700],
         'B': [500, 1000, 1500]}
df1 = pd.DataFrame(data1)

data2 = {'A': [70.90, 28.02, 4.003],
         'B': [500, 1000, 2500]}
df2 = pd.DataFrame(data2)

# Calcular medidas de tendencia central para el DataFrame 1
media_df1 = df1.mean()
mediana_df1 = df1.median()
moda_df1 = df1.mode().iloc[0]

# Calcular medidas de tendencia central para el DataFrame 2
media_df2 = df2.mean()
mediana_df2 = df2.median()
moda_df2 = df2.mode().iloc[0]

# Calcular medidas de dispersión para el DataFrame 1
rango_df1 = df1.max() - df1.min()
varianza_df1 = df1.var()
desviacion_estandar_df1 = df1.std()

# Calcular medidas de dispersión para el DataFrame 2
rango_df2 = df2.max() - df2.min()
varianza_df2 = df2.var()
desviacion_estandar_df2 = df2.std()

# Imprimir las medidas de tendencia central para ambos DataFrames
print("Medidas de tendencia central para DataFrame 1:")
print("Media:")
print(media_df1)
print("Mediana:")
print(mediana_df1)
print("Moda:")
print(moda_df1)

print("\nMedidas de tendencia central para DataFrame 2:")
print("Media:")
print(media_df2)
print("Mediana:")
print(mediana_df2)
print("Moda:")
print(moda_df2)

# Imprimir las medidas de dispersión para ambos DataFrames
print("\nMedidas de dispersión para DataFrame 1:")
print("Rango:")
print(rango_df1)
print("Varianza:")
print(varianza_df1)
print("Desviación Estándar:")
print(desviacion_estandar_df1)

print("\nMedidas de dispersión para DataFrame 2:")
print("Rango:")
print(rango_df2)
print("Varianza:")
print(varianza_df2)
print("Desviación Estándar:")
print(desviacion_estandar_df2)
```
